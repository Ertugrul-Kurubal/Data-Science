#### Groupby

```python
df_kargo_mean = df_kargo3.groupby("Ürün Adı")[["Kargodan alınan desi"]].mean()
df_kargo_mean.reset_index(inplace=True)
df_kargo_mean.rename(columns={"Kargodan alınan desi":"kargo desi ortalama"}, inplace=True)
df_kargo_mean["kargo desi ortalama"] = df_kargo_mean["kargo desi ortalama"].apply(lambda x: round(x,1))
df_kargo_mean
```

```python
df_kargo_max = df_kargo3.groupby("Ürün Adı")[["Kargodan alınan desi"]].max()
df_kargo_max.reset_index(inplace=True)
df_kargo_max.rename(columns={"Kargodan alınan desi":"kargo desi max"}, inplace=True)
df_kargo_max
```

```python
df_kargo_min = df_kargo3.groupby("Ürün Adı")[["Kargodan alınan desi"]].min()
df_kargo_min.reset_index(inplace=True)
df_kargo_min.rename(columns={"Kargodan alınan desi":"kargo desi min"}, inplace=True)
df_kargo_min
```

```python
df_kargo4 = df_kargo3.sort_values(by="Barkod")
df_kargo4 = df_kargo4.reset_index(drop=True)
df_kargo4.head(5)
```

```python
df_kargo4 = df_kargo4.drop_duplicates(subset=["Barkod","Ürün Adı","Kargo Firması"])
df_kargo4.reset_index(drop=True, inplace=True)
df_kargo4
```

```python
df_kargo_min_merge = pd.merge(df_kargo4, df_kargo_min, how="left", on="Ürün Adı")
df_kargo_min_merge
```

```python
df_kargo_min_max_merge = pd.merge(df_kargo_min_merge, df_kargo_max, how="left", on="Ürün Adı")
df_kargo_min_max_merge
```

```python
df_kargo_min_max_mean_merge = pd.merge(df_kargo_min_max_merge, df_kargo_mean, how="left", on="Ürün Adı")
df_kargo_min_max_mean_merge
```

```python
def kargo(var):
    '''
    This function converts cargo names.
    '''
    if var == "Trendyol Express Marketplace":
        return 'TEX'
    elif var == "MNG Kargo Marketplace":
        return 'MNG'
    elif var == 'Yurtiçi Kargo Marketplace':
        return 'Yurtiçi'
    elif var == 'PTT Kargo Marketplace':
        return 'PTT'
    elif var == 'Sürat Kargo Marketplace':
        return 'Sürat'
    else:
        return var
```

```python
df_trend_kargo["Kargo Firması"] = df_trend_kargo["Kargo Firması"].apply(lambda var: kargo(var))
df_trend_kargo
```

```python
df_trend_kargo2 = df_trend_kargo.dropna(subset="Kargodan alınan desi")
df_trend_kargo2
```
