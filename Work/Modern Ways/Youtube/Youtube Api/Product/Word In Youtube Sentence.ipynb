### Word In Youtube Sentence

```python
import pandas as pd
import numpy as np
import re
```

```python
word_num = 300
df_word_youtube = pd.read_csv("/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/Youtube/Result/Turkish/Word Tokenize Merge/Youtube_Word_Tokenize_Merge.csv")
df_word_select = df_word_youtube.head(word_num)
df_word_select
```

```python
df_sent_youtube = pd.read_csv("/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/Youtube/Result/Turkish/Sentence Clean Merge/Clean_Youtube_Sentence_Merge_Result.csv")
df_sent_youtube
```

```python
word_list = df_word_select.iloc[:,0].to_list()
```

```python
df_word_result = pd.DataFrame()
for i in word_list:
    try:
        word_in_video = df_sent_youtube[df_sent_youtube.sentence.str.contains(fr"(?:\s|^){i}(?:\s|$)", na=True)].sample(10) # try and except will try for sample
    except:
        word_in_video = df_sent_youtube[df_sent_youtube.sentence.str.contains(fr"(?:\s|^){i}(?:\s|$)", na=True)].head(10)         
    word_in_video.insert(0,"word",i)
    df_word_result = pd.concat([df_word_result,word_in_video], axis=0)
df_word_result.reset_index(drop=True, inplace=True)
df_word_result
```
