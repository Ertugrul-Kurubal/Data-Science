### Web Scrapping

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json 

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
from json2html import *
import re
```

#### From URL

```python
df = pd.read_excel("Videoid.xlsx") # Read Videoid excel file
df 
```

```python
df.id = df.id.apply(lambda x: x.strip())
df.id
```

```python
videoid = df.id.to_list()
videoid
```

```python
text_list = []
error_id = []
lang = "tr"
for i in videoid:
    try:
        url = f"https://www.youtube.com/api/timedtext?&v={i}&lang={lang}"
        html = urlopen(url)
        soup = BeautifulSoup(html)
        text_list.append(soup)
    except:
        error_id.append(i)
print(text_list)
```

```python
print(error_id)
```

#### Plain Text

```python
text_list = []
error_id = []
lang = "tr"
for i in videoid:
    try:
        url = f"https://www.youtube.com/api/timedtext?&v={i}&lang={lang}"
        html = urlopen(url)
        soup = BeautifulSoup(html)
        text_list.append(soup.text)
    except:
        error_id.append(i)
print(text_list)
```

```python
text = " ".join(text_list)
text
```

```python
with open("text.txt",'w',encoding = 'utf-8') as f:
   f.write(text)
```

#### Transcript Api

```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
video_list = ["oQyaI7Bt1Ig","0W-EUVhqv98"] # sample videoID
```

```python
video_sent = []
down_video_list = []
for i in video_list:
    try:
        var1 = YouTubeTranscriptApi.get_transcript(i)
        video_sent.append(var1)
        down_video_list.append(i)
    except:
        pass
video_sent
```

```python
down_video_list
```

```python
text_list = []
for i in range(len(video_sent)):
    for j in video_sent[i]:
        text_list.append(j)
df_list = pd.DataFrame(text_list)
df_list
```

```python
# Adding video_id
df_var = pd.DataFrame()
iter_item = zip(video_sent,down_video_list)
for video_text, id in iter_item:
    df_video = pd.DataFrame(video_text)
    df_video["video_id"] = id
    df_var = pd.concat([df_var, df_video], axis=0)
```
