### Compare DataFrame Columns

```python
import pandas as pd
import numpy as np
```

#### One Gram

```python
df_all_sent = pd.read_csv("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Pyspark/Result/Clean Spark Sentence Merge/Clean_Spark_Sentence_Merge_Result.csv")
df_all_sent = df_all_sent[["sentence","frequency"]]
df_all_sent
```

```python
df_one_gram = pd.read_csv("One_Gram_And_Sentence.csv") # one_gram and sentence
df_one_gram  
```

```python
#df_all_sent["match"] = np.where(df_all_sent.sentence == df_one_gram.sentence, True, False)
```

```python
set_all_sent = set(df_all_sent.iloc[:,0])
```

```python
set_one_gram = set(df_one_gram.iloc[:,0])
```

```python
set_more_one = set_all_sent.difference(set_one_gram)
```

```python
df_more_one = pd.DataFrame(set_more_one)
df_more_one.rename(columns={0:"sentence"}, inplace=True)
df_more_one
```

```python
df_more_one_word = pd.merge(df_all_sent, df_more_one, how='inner', on="sentence")
df_more_one_word
```

```python
df_more_one_word.to_csv("Sentence_More_Than_One_Word.csv", index=False)
```
