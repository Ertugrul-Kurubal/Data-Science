```python
# Two Gram
twogram_loc_list = []
for i in range(len(df_sent_twogram_result)):
    num_1 = df_sent_twogram_result.loc[i,"num_1"]
    num_2 = df_sent_twogram_result.loc[i,"num_2"]
    sent = df_sent_twogram_result.loc[i,"arabic_sent_tanzil_clean"]
    ngram = df_sent_twogram_result.loc[i,"twogram"]
    sent_token = word_tokenize(sent)
    ngram_token = word_tokenize(ngram)    
    for j in range(len(ngram_token)):        
        for k in range(len(sent_token)):
            if ngram_token[j] == sent_token[k]:
                try:
                    if ngram_token[j+1] == sent_token[k+1]:
                        num_3 = k+1  # +1 because of k begins with 0 
                        twogram_loc_list.append((num_1,num_2,num_3,ngram_token[j],sent))
                        twogram_loc_list.append((num_1,num_2,num_3+1,ngram_token[j+1],sent))
                except:
                    pass
                
df_twogram_loc = pd.DataFrame(twogram_loc_list)
df_twogram_loc.rename(columns={0:"num_1",1:"num_2",2:"num_3",3:"arabic_word_tanzil_clean",4:"arabic_sent_tanzil_clean"},inplace=True)
df_twogram_loc["first_200_twogram"] = 2
df_twogram_loc
```

```python
# Three Gram
df_sent_threegram_result = pd.DataFrame()
for i in df_threegram.iloc[:,0]:
    try:
        sent_in_video = df_sent[df_sent.arabic_sent_tanzil_clean.str.contains(fr"(^|\s){i}(\s|$)", regex=True)]
        #sent_in_video = df_sent[df_sent.arabic_sent_tanzil_clean.str.contains(fr"(^|\s){i}(\s|$)")].sample(10)
    except:
        sent_in_video = df_sent[df_sent.arabic_sent_tanzil_clean.str.contains(fr"(^|\s){i}(\s|$)", regex=True)] 
        #sent_in_video = df_sent[df_sent.arabic_sent_tanzil_clean.str.contains(fr"(^|\s){i}(\s|$)")].head(10) 
    sent_in_video.insert(0,"threegram",i)
    df_sent_threegram_result = pd.concat([df_sent_threegram_result,sent_in_video], axis=0)
df_sent_threegram_result.reset_index(drop=True, inplace=True)
df_sent_threegram_result
```

```python
# Three Gram
threegram_loc_list = []
for i in range(len(df_sent_threegram_result)):
    num_1 = df_sent_threegram_result.loc[i,"num_1"]
    num_2 = df_sent_threegram_result.loc[i,"num_2"]
    sent = df_sent_threegram_result.loc[i,"arabic_sent_tanzil_clean"]
    ngram = df_sent_threegram_result.loc[i,"threegram"]
    sent_token = word_tokenize(sent)
    ngram_token = word_tokenize(ngram)    
    for j in range(len(ngram_token)):        
        for k in range(len(sent_token)):
            if ngram_token[j] == sent_token[k]:
                try:
                    if ngram_token[j+1] == sent_token[k+1]:
                        if ngram_token[j+2] == sent_token[k+2]:
                            num_3 = k+1
                            threegram_loc_list.append((num_1,num_2,num_3,ngram_token[j],sent))
                            threegram_loc_list.append((num_1,num_2,num_3+1,ngram_token[j+1],sent))
                            threegram_loc_list.append((num_1,num_2,num_3+2,ngram_token[j+2],sent))
                except:
                    pass
                
df_threegram_loc = pd.DataFrame(threegram_loc_list)
df_threegram_loc.rename(columns={0:"num_1",1:"num_2",2:"num_3",3:"arabic_word_tanzil_clean",4:"arabic_sent_tanzil_clean"},inplace=True)
df_threegram_loc["100_200_threegram"] = 3
df_threegram_loc
```
