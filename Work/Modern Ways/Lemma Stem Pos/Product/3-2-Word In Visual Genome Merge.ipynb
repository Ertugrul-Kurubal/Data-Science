#### Word In Visual Genome Data

```python
df_objects_word = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Objects_Word_Result.csv")
df_objects_word.drop(["synsets","merged_object_ids"], inplace=True, axis=1)
df_objects_word_select = df_objects_word.drop(["object_id","height","width","x_koor","y_koor"], axis=1)
df_objects_word_select.rename(columns={"names":"search_text"}, inplace=True)
df_objects_word_select
```

```python
df_attributes_word = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Attributes_Word_Result.csv")
df_attributes_word.drop(["synsets"], inplace=True, axis=1)
df_attributes_word_select = df_attributes_word.drop(["object_id","names","height","width","x_koor","y_koor"], axis=1)
df_attributes_word_select.rename(columns={"attributes":"search_text"}, inplace=True)
df_attributes_word_select
```

```python
df_relationships_word = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Relationships_Word_Result.csv")
df_relationships_word.drop(["obj_names","obj_object_id","obj_synsets","obj_merged_object_ids","obj_height","obj_width","obj_x_koor","obj_y_koor","relationships_id","synsets","sub_synsets","predicate"], inplace=True, axis=1)
df_relationships_word.rename(columns={"sub_height":"height","sub_width":"width","sub_x_koor":"x_koor","sub_y_koor":"y_koor"}, inplace=True)
df_relationships_word_select = df_relationships_word.drop(["sub_object_id","height","width","x_koor","y_koor"], axis=1)
df_relationships_word_select.rename(columns={"sub_name":"search_text"}, inplace=True)
df_relationships_word_select
```

```python
df_question_answers_word = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Question_Answers_Word_Result.csv")
df_question_answers_word.drop("answers", inplace=True, axis=1)
df_question_answers_word_select = df_question_answers_word.drop("que_and_anw_id", axis=1)
df_question_answers_word_select.rename(columns={"questions":"search_text"}, inplace=True)
df_question_answers_word_select
```

```python
df_question_answers_word2 = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Question_Answers_Word_Result2.csv")
df_question_answers_word2.drop("questions", inplace=True, axis=1)
df_question_answers_word_select2 = df_question_answers_word2.drop("que_and_anw_id", axis=1)
df_question_answers_word_select2.rename(columns={"answers":"search_text"}, inplace=True)
df_question_answers_word_select2
```

```python
df_region_descriptions_word = pd.read_csv(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Result/3-1-Word In Visual Genome/Visual_Genome_Region_Descriptions_Word_Result.csv")
df_region_descriptions_word_select = df_region_descriptions_word.drop(["region_id","height","width","x_koor","y_koor"], axis=1)
df_region_descriptions_word_select.rename(columns={"phrases":"search_text"}, inplace=True)
df_region_descriptions_word_select
```

```python
df_genome_word_select = pd.concat([df_objects_word_select,df_attributes_word_select,df_relationships_word_select,df_question_answers_word_select,df_question_answers_word_select2,df_region_descriptions_word_select], axis=0)
df_genome_word_select.reset_index(drop=True, inplace=True)
df_genome_word_select
```

```python
df_genome_word_select["word_en_translate"].nunique()
```

```python
word_find_list = list(set(df_genome_word_select["word_en_translate"]))
```

```python
df_genome_word_select_raw_merge = pd.merge(df_word_raw, df_genome_word_select, how="inner", on="word_en_translate")
df_genome_word_select_raw_merge
```

```python
df_genome_word_select_image_count = df_genome_word_select_raw_merge.groupby(["word"])["image_id"].count().reset_index()
df_genome_word_select_image_count.rename(columns={"image_id":"image_id_count"}, inplace=True)
df_genome_word_select_image_count.sort_values("image_id_count", ascending=False, inplace=True)
df_genome_word_select_image_count
```

```python
df_genome_word_all = pd.concat([df_objects_word,df_attributes_word,df_relationships_word,df_question_answers_word,df_question_answers_word2,df_region_descriptions_word], axis=0)
df_genome_word_all.reset_index(drop=True, inplace=True)
df_genome_word_all
```

```python
df_genome_word_all_raw_merge = pd.merge(df_word_raw, df_genome_word_all, how="inner", on="word_en_translate")
df_genome_word_all_raw_merge
```

```python
df_genome_word_all_image_count = df_genome_word_all_raw_merge.groupby(["word"])["image_id"].count().reset_index()
df_genome_word_all_image_count.rename(columns={"image_id":"image_id_count"}, inplace=True)
df_genome_word_all_image_count.sort_values("image_id_count", ascending=False, inplace=True)
df_genome_word_all_image_count
```
