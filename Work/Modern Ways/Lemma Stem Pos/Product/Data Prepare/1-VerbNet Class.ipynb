```python
def lower_strip_func(x):
    try:
        var_low = str(x).lower()
        var_out = var_low.strip()
    except:
        var_out = x
    return var_out 
```

#### Verb Class Verbnet Merge

```python
df_verb_class = pd.read_excel(f"/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/{lang_folder.capitalize()}/Lemma Stem POS/Data/VerbNet/VerbNet_Class.xlsx")
df_verb_class
```

```python
df_verb_class["number"] = df_verb_class["number"].apply(lambda x: lower_strip_func(x))
df_verb_class["class"] = df_verb_class["class"].apply(lambda x: lower_strip_func(x))
```

```python
df_verb_verbnet = pd.read_excel(f"/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/{lang_folder.capitalize()}/Lemma Stem POS/Data/VerbNet/VerbNet_Verbnet.xlsx")
df_verb_verbnet
```

```python
df_verb_verbnet["number"] = df_verb_verbnet["number"].apply(lambda x: lower_strip_func(x))
df_verb_verbnet["verbnet"] = df_verb_verbnet["verbnet"].apply(lambda x: lower_strip_func(x))
```

```python
df_verb_class_verbnet_merge = pd.merge(df_verb_class,df_verb_verbnet,how="inner", on="number")
df_verb_class_verbnet_merge
```

```python
df_verb_class_verbnet_merge.to_excel(f"Verb_Class_Verbnet_Merge.xlsx", index=False)
```

#### Read XML File

```python
xml_source_file_path = "/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/Turkish/Lemma Stem POS/Data/VerbNet/new_vn"
```

```python
xml_file_list = os.listdir(xml_source_file_path)
xml_file_list
```

```python
verb_list = []
number_list = []
name_list = []
wn_list = []
for xml_file in xml_file_list:
    file_without_ext, number = os.path.splitext(xml_file)[0].split("-")
    try:
        df_var = pd.read_xml(f"{xml_source_file_path}/{xml_file}",xpath=".//MEMBER")
        zip_list = zip(df_var["name"],df_var ["wn"])
        for name, wn in zip_list:
            verb_list.append(file_without_ext)
            number_list.append(number)
            name_list.append(name)
            wn_list.append(wn)
    except:
        pass
    
df_verb_class_verbnet = pd.DataFrame({"number":number_list,"class":verb_list,"verbnet":name_list,"wordnet":wn_list})
df_verb_class_verbnet
```

```python
df_verb_class_verbnet[["verb","wn_number"]] = df_verb_class_verbnet["wordnet"].str.split("%", 1, expand=True)
df_verb_class_verbnet
```

```python
df_verb_class_verbnet.to_excel(f"Verb_Class_Verbnet_XML.xlsx", index=False)
```
