#### Selected Image Rename

```python
# funtion for remove brackets and number from filename
def clean_text(text):
    text2 = re.sub('[\(\[\{].*?[\)\]\}]', '', text)
    text3 = re.sub('[0-9]', '', text2)
    text4 = text3.strip()
    return text4
```

```python
# source destination path folders
source_path_folder = output_path_folder
destination_path_folder = f"{output_path}/Genome_Picture_In_Picture_Select_Rename"

Path(destination_path_folder).mkdir(parents=True, exist_ok=True)
```

```python
file_list = os.listdir(source_path_folder)
file_list
```

```python
for file in file_list:
    # file and extention
    file_without_ext = os.path.splitext(file)[0]
    file_ext = os.path.splitext(file)[1]
    # file clean and join
    file_without_ext_clean = clean_text(file_without_ext)
    word_list = word_tokenize(file_without_ext_clean)
    file_without_ext_new = " ".join(word_list)
    file_new = f"{file_without_ext_new.lower()}{file_ext}"
    # check move rename delete
    if isfile(f"{source_path_folder}/{file_new}"):
        pass
    else:
        source_file = f"{source_path_folder}/{file}"
        destination_path = f"{destination_path_folder}"        
        shutil.copy2(source_file, destination_path)
        os.remove(source_file)
        source_dest_file = f"{destination_path}/{file}"
        target_dest_file = f"{destination_path}/{file_new}"
        os.rename(source_dest_file, target_dest_file)

os.rmdir(source_path_folder)  # delete empty folder
```
