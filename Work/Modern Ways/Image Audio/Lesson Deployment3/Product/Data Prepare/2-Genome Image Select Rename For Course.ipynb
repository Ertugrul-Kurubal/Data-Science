#### Selected Image Rename

```python
# funtion for remove brackets and number from filename
def clean_text(text):
    text2 = re.sub('[\(\[\{].*?[\)\]\}]', '', text)
    text3 = re.sub('[0-9]', '', text2)
    text4 = text3.strip()
    return text4
```

```python
# source destination path folders
source_path_folder = f"{output_path}/Genome_Picture_In_Picture_{folder_extention}"
destination_path_folder = f"{output_path}/Genome_Picture_In_Picture_Rename_{folder_extention}"

Path(destination_path_folder).mkdir(parents=True, exist_ok=True)
```

```python
file_list = os.listdir(source_path_folder)
file_list
```

```python
for file in file_list:
    # file and extention
    file_without_ext = os.path.splitext(file)[0]
    file_ext = os.path.splitext(file)[1]
    # file clean and join
    file_without_ext_clean = clean_text(file_without_ext)
    word_list = word_tokenize(file_without_ext_clean)
    file_without_ext_new = " ".join(word_list)
    file_new = f"{file_without_ext_new.lower()}{file_ext}"
    # check move rename delete
    if isfile(f"{source_path_folder}/{file_new}"):
        pass
    else:
        source_file = f"{source_path_folder}/{file}"
        destination_path = f"{destination_path_folder}"        
        shutil.copy2(source_file, destination_path)
        os.remove(source_file)
        source_dest_file = f"{destination_path}/{file}"
        target_dest_file = f"{destination_path}/{file_new}"
        os.rename(source_dest_file, target_dest_file)

os.rmdir(source_path_folder)  # delete empty folder
```

#### Select Image According To Max Size

```python
path_folder_list = glob.glob(f"{image_path}/Genome_Picture_In_Picture/*")
path_folder_list
```

```python
selected_image_file_list = []
for path_folder in path_folder_list:
    image_files = glob.glob(f"{path_folder}/*")
    #image_file_var = image_files[0]  # select adjust image
    image_file_var = max(image_files, key=lambda x: os.stat(x).st_size)  # select max size file
    selected_image_file_list.append(image_file_var)
selected_image_file_list
```

```python
output_folder = "Genome_Picture_In_Picture_Select"
output_path_folder = f"{output_path}/{output_folder}"

Path(output_path_folder).mkdir(parents=True, exist_ok=True)
```

```python
for l in selected_image_file_list:
    source = l # source directory
    destination = output_path_folder
    shutil.copy2(source, destination)
```
