```python
df_github_cross = pd.read_csv("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Web Scrapping/Data/All/etymology_cross_check2.csv")
df_github_cross
```

```python
lang_1 = "Turkish"  # English, Turkish, German, French, Spanish, Arabic, Russian, Portuguese, Dutch, Italian
lang_2 = "French"  # English, Turkish, German, French, Spanish, Arabic, Russian, Portuguese, Dutch, Italian
```

```python
def low_func(x):
    try:
        var_out = x.lower()
    except:
        var_out = x
    return var_out 
```

```python
df_lang1 = df_github_cross[(df_github_cross["lang"] == f"{lang_1.capitalize()}")] 
df_lang1
```

```python
#df_lang1.to_excel("Turkish_Word_From_Other_Language.xlsx", index=False)
```

```python
df_lang2 = df_github_cross[(df_github_cross["lang"] == f"{lang_2.capitalize()}")]
df_lang2["term"] = df_lang2["term"].apply(lambda x: x.lower())
#df_lang2.rename(columns={"term":"word"}, inplace=True)
df_lang2
```

```python
#df_word = pd.read_excel(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/{lang_2.capitalize()}/Result/Word/Merge/Word_Merge2.xlsx")
#df_word = df_word.head(40000)
#df_word = df_word.loc[:,["word"]]
#df_word
```

```python
#df_lang2 = pd.merge(df_lang2,df_word,how="inner",on="word")
#df_lang2
```

```python
set_lang1 = set(df_lang1["related_term_id"])
set_lang2= set(df_lang2["related_term_id"])
```

```python
df_pair_etymology = pd.DataFrame(set_lang1.intersection(set_lang2),columns=["related_term_id"])
df_pair_etymology
```

```python
df_etymology_merge = pd.merge(df_github_cross,df_pair_etymology,how="inner",on="related_term_id")
df_etymology_merge.drop_duplicates(inplace=True)
df_etymology_merge
```

```python
df_etymology_merge = df_etymology_merge[(df_etymology_merge["reltype"] == "inherited_from") | (df_etymology_merge["reltype"] == "etymologically_related_to") | (df_etymology_merge["reltype"] == "derived_from") | (df_etymology_merge["reltype"] == "borrowed_from")]
df_etymology_merge
```

```python
df_etymology_merge = df_etymology_merge[~df_etymology_merge["related_term"].str.contains(fr"(?:\s|^)\*|\-(?:\s|$)", na=True)]
df_etymology_merge
```

```python
df_lang1_result = df_etymology_merge[(df_etymology_merge["lang"]==f"{lang_1.capitalize()}")]
df_lang1_result.drop_duplicates(inplace=True)
df_lang1_result.reset_index(drop=True, inplace=True)
df_lang1_result
```

```python
df_lang1_result["term"] = df_lang1_result["term"].apply(lambda x: low_func(x))
df_lang1_result["related_term"] = df_lang1_result["related_term"].apply(lambda x: low_func(x))
df_lang1_result = df_lang1_result.loc[:,["lang","term","related_term_id"]]
df_lang1_result
```

```python
df_lang2_result = df_etymology_merge[(df_etymology_merge["lang"]==f"{lang_2.capitalize()}")]
df_lang2_result.drop_duplicates(inplace=True)
df_lang2_result.reset_index(drop=True, inplace=True)
df_lang2_result
```

```python
df_lang2_result["term"] = df_lang2_result["term"].apply(lambda x: low_func(x))
df_lang2_result["related_term"] = df_lang2_result["related_term"].apply(lambda x: low_func(x))
df_lang2_result = df_lang2_result.loc[:,["lang","term","related_term_id"]]
df_lang2_result
```

```python
df_same_ety = pd.merge(df_lang1_result,df_lang2_result, how="inner", on="related_term_id", suffixes=["_master","_sleeve"])
df_same_ety = df_same_ety[["lang_master","term_master","lang_sleeve","term_sleeve"]]
df_same_ety.drop_duplicates(inplace=True)
df_same_ety
```

```python
df_same_ety.to_excel(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Web Scrapping/Result/{lang_1.capitalize()}/{lang_1.capitalize()} {lang_2.capitalize()}/{lang_1.capitalize()}_And_{lang_2.capitalize()}_Etymologeek_Result2.xlsx", index=False)
```
