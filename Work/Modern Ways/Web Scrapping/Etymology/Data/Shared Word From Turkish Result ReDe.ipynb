```python
# custom character alphabet for word
ar = re.compile(r"[ٿصؼۤڳڲؿڎػڠجڿ٬ٸؽؒؓطۄڀۂؘؔتٚڛےٝڜؖڦ٫ډ۰زۇٖۀ،لۓعٮێڔ۶ؚۧۜڤۏإٞٷؗۖ؈ژۣؕؑٴأۻڸۺگاڴڹۯ؉ْڌ؍ي؟ـٟړۅؐڶُىڽېًۢؠضۚڄٛڏٱۦ٩س٦ڼڂٔۘ٠ښٌٍ۬ٳ۾ٲږذۋٵٜ٘ڞڅںٗهڣۿپڒۥۗڋیؙم؞ثۨٹڵڪظٶۭ١ڭەڨحٕ؎ٺڷٰ۪۫ڻڥۛڑڟټآڡغګ؊ّٯڧڮ؏ۮ؋ؤ٪ؾڗۼق۟دکوِڰڐۃ۽ہفرڇچڝ۴بۈٽڕۡھةٓڃئ؛ڬٙڙڢڱۊَۆۉځ۠ۍۑۙڊنءڈٻشڍ؇۵كخ\ا/]+", re.IGNORECASE|re.UNICODE) # Arabic
```

```python
def clean_text(text,custom_alp):
    #text_clean = re.findall(custom_alp, text)
    text_clean = re.findall(custom_alp, str(text))
    text_result = " ".join(text_clean)
    #text_result_list = re.findall(r"\w+",text_result, re.IGNORECASE|re.UNICODE)
    return text_result
```

```python
def lower_strip_func(x):
    try:
        var_low = x.lower()
        var_out1 = var_low.strip()
        var_out2 = var_out1.strip("//")
    except:
        var_out2 = x
    return var_out2 
```

```python
df_arabic_word = pd.read_excel("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Web Scrapping/Data/All/Arapca Kelimeler.xlsx", sheet_name="arapca")
df_arabic_word
```

```python
df_arabic_word["len"] = df_arabic_word["koken"].apply(lambda x: len(x))
df_arabic_word
```

```python
len_mean = df_arabic_word["len"].mean()
len_mean
```

```python
def len_select(string,start,end):
    try:
        string = string[start:end]
    except:
        string = string
    return string
```

```python
df_arabic_word["koken"] = df_arabic_word["koken"].apply(lambda x: len_select(x,0,len_mean))
```

```python
df_arabic_word
```

```python
df_arabic_word.loc[:,"koken"] = df_arabic_word.loc[:,"koken"].apply(lambda x: clean_text(x, ar))
df_arabic_word.dropna(inplace=True)
df_arabic_word.drop(df_arabic_word[df_arabic_word.iloc[:,0] == ""].index, inplace=True)
df_arabic_word["word"] = df_arabic_word["word"].apply(lambda x: lower_strip_func(x))
df_arabic_word["koken"] = df_arabic_word["koken"].apply(lambda x: lower_strip_func(x))
df_arabic_word.drop_duplicates(subset="word", inplace=True)
df_arabic_word.reset_index(drop=True, inplace=True)
#df_arabic_word.drop("len", axis=1, inplace=True)
df_arabic_word
```
