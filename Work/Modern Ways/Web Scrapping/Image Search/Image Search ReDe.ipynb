```python
from icrawler.builtin import BaiduImageCrawler, BingImageCrawler, GoogleImageCrawler
```

```python
# Google
#from icrawler.builtin import GoogleImageCrawler

google_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=2,
    downloader_threads=4,
    storage={'root_dir': 'images'})
filters = dict(
    size='large',
    license='commercial,modify',
    date=((2020, 1, 1), (2020, 11, 30)))
google_crawler.crawl(keyword='cat', filters=filters, max_num=10, file_idx_offset=0)
```

```python
# Bing
bing_crawler = BingImageCrawler(downloader_threads=4,
                                storage={'root_dir': 'images'})
bing_crawler.crawl(keyword='cat', filters=None, offset=0, max_num=10)
```

```python
# language pair
lang_folder = "English"  # Arabic, English, French, German, Turkish, Spanish, Portuguese, Dutch, Italian ==> target language for learner
```

```python
df_word_all = pd.read_excel(f"/media/kurubal/SSD1/Data Scientist/Work/Modern Ways/Project/{lang_folder.lower().capitalize()}/Deployment/Data/Word/Word_Merge_Preprocess.xlsx")
df_word_all = df_word_all.head(28)
df_word_all
```

```python
word_list = df_word_all["word"].to_list()
```

```python
from icrawler.builtin import GoogleImageCrawler

for keyword in word_list:
    google_crawler = GoogleImageCrawler(
        parser_threads=2,
        downloader_threads=4,
        storage={'root_dir': 'google_images/{}'.format(keyword)}
    )
    google_crawler.crawl(
        keyword=keyword, max_num=5, min_size=(500, 500))
```

```python
for keyword in word_list:
    bing_crawler = BingImageCrawler(downloader_threads=4,
                                storage={'root_dir': 'bing_images/{}'.format(keyword)})
    bing_crawler.crawl(keyword=keyword, filters=None, offset=0, max_num=5)
```
