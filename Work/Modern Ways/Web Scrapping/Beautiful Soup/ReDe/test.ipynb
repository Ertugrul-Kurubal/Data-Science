```python
df_github_cross = pd.read_csv("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Web Scrapping/Data/All/etymology_cross_check2.csv")
#df_github_cross = df_github_cross.head(150000)
#df_github_cross = df_github_cross.loc[:,["lang","term","related_lang","related_term"]]
df_github_cross
```

```python
lang_1 = "Turkish"  # English, Turkish, German, French, Spanish, Arabic, Russian, Portuguese, Dutch, Italian
lang_2 = "English"  # English, Turkish, German, French, Spanish, Arabic, Russian, Portuguese, Dutch, Italian
```

```python
df_lang1 = df_github_cross[(df_github_cross["lang"] == f"{lang_1.capitalize()}")] 
df_lang1
```

```python
df_lang2 = df_github_cross[(df_github_cross["lang"] == f"{lang_2.capitalize()}")]
df_lang2
```

```python
set_lang1 = set(df_lang1["related_term_id"])
set_lang2= set(df_lang2["related_term_id"])
```

```python
df_root = pd.DataFrame(set_lang1.intersection(set_lang2),columns=["related_term_id"])
df_root
```

```python
df_root_merge = pd.merge(df_github_cross,df_root,how="inner",on="related_term_id")
df_root_merge.drop_duplicates(inplace=True)
df_root_merge
```

```python
test_main = df_root_merge[(df_root_merge["lang"]==f"{lang_1.capitalize()}")]
test_main.drop_duplicates(inplace=True)
test_main
```

```python
test_pair = df_root_merge[(df_root_merge["lang"]==f"{lang_2.capitalize()}")]
test_pair.drop_duplicates(inplace=True)
test_pair
```

```python
df_result = pd.merge(test_main, test_pair, how="inner",on="related_term")
df_result.drop_duplicates(inplace=True)
df_result
```

```python
#df_result = df_result.head(1000000)
```

```python
df_result.to_excel(f"{lang_1.capitalize()}_And_{lang_2.capitalize()}_Same_Root.xlsx",index=False)
```
