```python
import pandas as pd
import numpy as np
import json
```

```python
df_json = pd.read_json("output.json").transpose()
df_json.reset_index(drop=True, inplace=True)
df_json = df_json.loc[:,["baslik", "koken"]]
df_json.rename(columns={"baslik":"word"}, inplace=True)
df_json
```

```python
df_json["word"] = df_json["word"].apply(lambda x: x.strip())
```

```python
df_word = pd.read_excel(f"/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Turkish/Result/Word/Merge/Word_Merge2.xlsx")
df_word = df_word.head(40000)
df_word = df_word.loc[:,["word"]]
df_word
```

```python
set_json = set(df_json.word)
set_word = set(df_word.word)
```

```python
len(set_json.intersection(set_word))
```

```python
df_intersect = pd.DataFrame(set_json.intersection(set_word), columns=["intersect_word"])
df_intersect
```

```python
df_merge = pd.merge(df_word,df_intersect,how="left", left_on="word",right_on="intersect_word")
df_merge.rename(columns={"intersect_word":"nisanyan"}, inplace=True)
df_merge
```

```python
df_merge['nisanyan'][~df_merge['nisanyan'].isna()] = "nisanyan"  # word convert nisanyan 
```
