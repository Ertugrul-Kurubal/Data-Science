```python
#df_twogram = pd.read_excel("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Arabic/Quran/Text/Result/4-N Gram/Arabic_Clean_Twogram_Frequency_From_DataFrame_Tanzil.xlsx")
df_twogram = pd.read_excel("/media/kurubal/SSD/Data Scientist/Work/Modern Ways/Project/Arabic/Quran/Text/Result/4-N Gram/Arabic_Clean_Twogram_Frequency_From_DataFrame_Tanzil2.xlsx")
df_twogram_select = df_twogram.head(200)
df_twogram_select
```

```python
def twogram_loc_detect(df_twogram, df_target, ngram_column_name, target_column_name): # df_twogram is selected twogram words dataframe, ngram_column_name acquires from target_column_name
    for i in range(len(df_twogram)):
        twogram = df_twogram.loc[i,f"{ngram_column_name}"]
        twogram_words = word_tokenize(twogram)
        word_1 = twogram_words[0]
        word_2 = twogram_words[1]
        for j in range(len(df_target)):
            try:
                target_word_1 = df_target.loc[j,f"{target_column_name}"]
                target_word_2 = df_target.loc[j+1,f"{target_column_name}"]
                if (word_1 == target_word_1) and (word_2 == target_word_2):
                    df_target.loc[j,"twogram_string"] = f"{twogram}"
                    #df_target.loc[j+1,"twogram_string"] = f"{twogram}"                    
                else:
                    pass
            except:
                pass
    df_result = df_target
           
    return df_result
```

#### Detect Multi Twogram Indexes

```python
def twogram_loc_index_detect(df_twogram, df_target, ngram_column_name, target_column_name, index_column_name):
    for i in range(len(df_twogram)):
        twogram = df_twogram.loc[i,f"{ngram_column_name}"]
        twogram_words = word_tokenize(twogram)
        word_1 = twogram_words[0]
        word_2 = twogram_words[1]
        for j in range(len(df_target)):
            try:
                target_word_1 = df_target.loc[j,f"{target_column_name}"]
                target_word_2 = df_target.loc[j+1,f"{target_column_name}"]
                index_1 = df_target.loc[j,f"{index_column_name}"]
                index_2 = df_target.loc[j+1,f"{index_column_name}"]
                #variable_value = [variable_1, variable_2]
                if (word_1 == target_word_1) and (word_2 == target_word_2):
                    df_target.loc[j,"twogram_string"] = f"{twogram}"
                    df_target.loc[j,"first_index"] = index_1
                    df_target.loc[j,"second_index"] = index_2                    
                else:
                    pass
            except:
                pass
    df_result = df_target

    return df_result
```
