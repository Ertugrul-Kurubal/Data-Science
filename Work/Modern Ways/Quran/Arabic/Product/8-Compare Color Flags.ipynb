```python
df = pd.merge(df_all_color_flag, df_all_level_flag, how="inner", on=["num_1","num_2","num_3","index"])
df.drop_duplicates(inplace=True)
df
```

```python
# df first column order must be (num_1, num_2, num_3, arabic_word_tanzil_clean, buckwalter_word_tanzil_clean)
for i in range(len(df)):
    num_2 = df.iloc[i,1] # num_2
    num_3 = df.iloc[i,2] # num_3
    buckwalter_word = df.iloc[i,4] # buckwalter_word_tanzil_clean
    if num_2 == 1:
        if (buckwalter_word == "bsm" and df.iloc[i+1,4] == "Allh" and df.iloc[i+2,4] == "AlrHmn")  or (df.iloc[i-1,4] == "bsm" and buckwalter_word == "Allh" and df.iloc[i+1,4] == "AlrHmn") \
            or (df.iloc[i-1,4] == "Allh" and buckwalter_word == "AlrHmn" and df.iloc[i+1,4] == "AlrHym") or (df.iloc[i-2,4] == "Allh" and  df.iloc[i-1,4] == "AlrHmn" and buckwalter_word == "AlrHym"):
            df.iloc[i,1] = 0
        elif num_3 == 1:
            pass
        elif num_3 != 1 and df.iloc[i-1,2] == 4 and df.iloc[i-1,4] == "AlrHym":
            df.iloc[i,2] = 1
        else:
            df.iloc[i,2] = df.iloc[i-1,2]+1
    else:
        pass
```

```python
# df first column order must be (num_1, num_2, num_3, arabic_word_tanzil_clean, buckwalter_word_tanzil_clean)
def relocation(df):
    for i in range(len(df)):
        num_2 = df.iloc[i,1] # num_2
        num_3 = df.iloc[i,2] # num_3
        buckwalter_word = df.iloc[i,4] # buckwalter_word_tanzil_clean
        if num_2 == 1:
            if (buckwalter_word == "bsm" and df.iloc[i+1,4] == "Allh" and df.iloc[i+2,4] == "AlrHmn")  or (df.iloc[i-1,4] == "bsm" and buckwalter_word == "Allh" and df.iloc[i+1,4] == "AlrHmn") \
                or (df.iloc[i-1,4] == "Allh" and buckwalter_word == "AlrHmn" and df.iloc[i+1,4] == "AlrHym") or (df.iloc[i-2,4] == "Allh" and  df.iloc[i-1,4] == "AlrHmn" and buckwalter_word == "AlrHym"):
                df.iloc[i,1] = 0
            elif num_3 == 1:
                pass
            elif num_3 != 1 and df.iloc[i-1,2] == 4 and df.iloc[i-1,4] == "AlrHym":
                df.iloc[i,2] = 1
            else:
                df.iloc[i,2] = df.iloc[i-1,2]+1
        else:
            pass
    df.rename(columns={"num_1":"surah","num_2":"ayah","num_3":"word rank","arabic_word_tanzil_clean":"tanzil_clean","arabic_word_tanzil_plain":"tanzil_plain"},inplace=True)
    return df
```
