```python
# N gram from text data and DataFrame output
#ngram_type = "twogram"
def text_ngram(text, ngram_num, ngram_type): # text_ngram(arabic_concat, 2, "twogram") ...
    n_gram = ngrams(text.split(), ngram_num)
    n_gram_list = []
    for i in n_gram:
        n_gram_group = " ".join(i)
        n_gram_list.append(n_gram_group)
    df_ngram = pd.DataFrame(n_gram_list, columns=[f"{ngram_type.lower()}"])
    df_ngram = pd.DataFrame(df_ngram.iloc[:,0].value_counts(ascending=False))
    df_ngram.reset_index(inplace=True)
    df_ngram.rename(columns={"index":f"{ngram_type.lower()}", f"{ngram_type.lower()}":"frequency"}, inplace=True)
    return df_ngram
```

```python
# N gram from DataFrame text and DataFrame output
def df_ngram(df_column, ngram_num, ngram_type): # df_ngram(df_buckwalter_concat.iloc[:,2], 2, "twogram") ...
    n_gram_not_list = []
    n_gram_list = []
    for i in df_column:
        try:
            n_gram = ngrams(i.split(), ngram_num)
            for j in n_gram:
                j = " ".join(j)
                n_gram_list.append(j)
        except:
            n_gram_not_list.append(i) 
    df_ngram = pd.DataFrame(n_gram_list,columns=[f"{ngram_type}"])
    df_ngram = df_ngram.iloc[:,0].value_counts(ascending=False).reset_index()
    df_ngram.rename(columns={"index":f"{ngram_type.lower()}", f"{ngram_type.lower()}":"frequency"}, inplace=True)
    return df_ngram
```
