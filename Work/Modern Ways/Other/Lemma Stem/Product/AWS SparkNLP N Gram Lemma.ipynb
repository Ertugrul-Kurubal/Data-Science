### Lemma Model Result

```python
df_word = pd.read_excel("10000 word.xlsx")
df_word
```

```python
df_lemma = pd.read_json("tr.lemma.atis.json")
df_lemma
```

```python
df_lemma = df_lemma[["token","lem"]]
df_lemma.rename(columns={"token":"word", "lem":"lemma.atis"}, inplace=True)
df_lemma
```

```python
df_word_lemma_merge = pd.merge(df_word, df_lemma, how="left", on="word")
df_word_lemma_merge.drop_duplicates(inplace=True)
df_word_lemma_merge
```

```python
df_lemma = pd.read_json("tr.lemma.tourism.json")
df_lemma = df_lemma[["token","lem"]]
df_lemma.rename(columns={"token":"word", "lem":"lemma.tourism"}, inplace=True)
df_lemma
```

```python
df_word_lemma_merge = pd.merge(df_word_lemma_merge, df_lemma, how="left", on="word")
df_word_lemma_merge.drop_duplicates(inplace=True)
df_word_lemma_merge
```

```python
df_word_lemma_merge.to_excel("Turkish_10000_Word_All_Lemma_Result.xlsx", index=False)
```
