```python
df_word = pd.DataFrame()
for i in range(len(result_data)):
    data = result_data[i]["alternatives"]
    df_data = pd.json_normalize(data,record_path=["words"])
    df_word = pd.concat([df_word,df_data], axis=0) # attention df_word,df_data must be for ascending time 
df_word.reset_index(drop=True, inplace=True)
df_word 
```
