```python
df.tarih = pd.to_datetime(df.tarih)
```

```python
def get_date_int(df, column):
    year_1 = df[column].dt.year
    month_1 = df[column].dt.month
    day_1 = df[column].dt.day
    return year_1, month_1, day_1
```

```python
df["yıl"], df["ay"], df["gün"] = get_date_int(df, "tarih") 
```

```python
df.drop("tarih",axis=1, inplace=True)
```
